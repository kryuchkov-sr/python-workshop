```python
# Задание task_02_02_02
# Выполнил: Крючков С. Р.
# Группа: АБП-231

import random

def generate_random_int(min_val, max_val):
    """Сгенерировать случайное целое число в диапазоне [min_val, max_val].

    Параметры:
        - min_val (int): минимальное значение диапазона.
        - max_val (int): максимальное значение диапазона.

    Результат:
        - int: случайное целое число."""

    random_1_100 = random.randint(min_val, max_val)
    return random_1_100

min_val = 1
max_val = 100
print(f"Случайное число от {min_val} до {max_val}: {generate_random_int(min_val, max_val)}")

# --------------
# Пример вывода:
#
# Случайное число от 1 до 100: 42  (значение может отличаться)
```

```python
# Задание task_02_02_02
# Выполнил: Крючков С. Р.
# Группа: АБП-231

import math

def hypotenuse(a, b):
  """Вычислить гипотенузу по двум катетам 'a' и 'b'.
    Параметры:
        - a (float): длина первого катета.
        - b (float): длина второго катета.
    Результат:
        - float: длина гипотенузы."""

  gipotenuza = math.sqrt(a ** 2 + b ** 2)
  return gipotenuza


cathetus1 = 3.0
cathetus2 = 4.0
print(f"Гипотенуза для катетов {cathetus1} и {cathetus2}: {hypotenuse(cathetus1, cathetus2):.2f}")

# --------------
# Пример вывода:
#
# Гипотенуза для катетов 3.0 и 4.0: 5.00
```

```python
# Задание task_02_02_06
# Выполнил: Крючков С. Р.
# Группа: АБП-231

import math

def sin_of_angle(angle_rad):
  """Вычислить синус угла, заданного в радианах.
    Параметры:
        - angle_rad (float): угол в радианах.
    Результат:
        - float: синус угла."""

  sinus = math.sin(angle_rad)
  return sinus

angle = (math.pi / 2)  # 90 градусов в радианах
print(f"Синус угла в {angle:.2f} радиан: {sin_of_angle(angle):.2f}")

# --------------
# Пример вывода:
#
# Синус угла 1.57 радиан: 1.00
```

```python
# Задание task_02_02_02
# Выполнил: Крючков С. Р.
# Группа: АБП-231

import random

def shuffle_list(input_list):
  """Перемешать элементы списка 'input_list' на месте.
    Параметры:
        - input_list (list): список для перемешивания.
    Результат:
        - None: функция изменяет список на месте."""

  random.shuffle(input_list)

my_list = [1, 2, 3, 4, 5]
print(f"Исходный список: {my_list}")
shuffle_list(my_list)
print(f"Перемешанный список: {my_list}")

# --------------
# Пример вывода:
#
# Исходный список: [1, 2, 3, 4, 5]
# Перемешанный список: [3, 5, 1, 2, 4] (порядок может отличаться)
```

```python
# Задание task_02_02_08
# Выполнил: Крючков С. Р.
# Группа: АБП-231

import datetime

def days_between_dates(date1, date2):
  """Вычислить количество дней между двумя датами.
    Параметры:
        - date1 (datetime.date): первая дата.
        - date2 (datetime.date): вторая дата.
    Результат:
        - int: количество дней между датами."""

  delta = date2 - date1
  return delta

date_start = datetime.date(2023, 1, 15)
date_end = datetime.date(2023, 2, 10)
print(f"Количество дней между {date_start} и {date_end}: {days_between_dates(date_start, date_end)}")

# --------------
# Пример вывода:
#
# Количество дней между 2023-01-15 и 2023-02-10: 26
```

```python
# Задание task_02_02_09
# Выполнил: Крючков С. Р.
# Группа: АБП-231

import math

def integer_part(number):
  """Получить целую часть числа, отбросив дробную.
    Параметры:
        - number (float): число.
    Результат:
        - int: целая часть числа."""

  if number * 10 % 10 > 5:
    return math.ceil(number)
  else:
    return math.floor(number)

number = 123.456
print(f"Целая часть числа {number}: {integer_part(number)}")

# --------------
# Пример вывода:
#
# Целая часть числа 123.456: 123
```

```python
# Задание task_02_02_02
# Выполнил: Крючков С. Р.
# Группа: АБП-231

import random

def choose_random_element(input_list):
  """Выбрать один случайный элемент из списка.
  Параметры:
        - input_list (list): непустой список.
  Результат:
        - any: случайный элемент из списка."""

  choice_element = random.choice(input_list)
  return choice_element

my_list = ["apple", "banana", "cherry", "date"]
print(f"Список: {my_list}")
print(f"Случайный элемент: {choose_random_element(my_list)}")

# --------------
# Пример вывода:
#
# Список: ['apple', 'banana', 'cherry', 'date']
# Случайный элемент: cherry (значение может отличаться)
```
