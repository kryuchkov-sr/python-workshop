```python
#  task_02_34
# :  
# : -231

# #      ,
# #     
# #     
# #     
# deposits = [
#     {"name": "_1", "initial_sum": 50000, "rate": 5.2},
#     {"name": "_2", "initial_sum": 20000, "rate": 3.7},
#     {"name": "_3", "initial_sum": 45000, "rate": 6.8},
# ]

# 1.  
n = int(input("  = "))

deposits = []
for i in range(n):
    line = input().split()
    name = line[0]
    initial_sum = int(line[1])
    rate = float(line[2])
    deposits.append({
        "name": name,
        "initial_sum": initial_sum,
        "rate": rate
    })

#   
print(f"  = {n}")
for dep in deposits:
    print(f"{dep['name']} {dep['initial_sum']} {dep['rate']}")

# 2.    (  )
min_bank = min(deposits, key=lambda x: x["initial_sum"])
print(f"  : {min_bank['name']},  : {min_bank['initial_sum']:.2f} .")

# 3.    ( )
max_profit_bank = max(deposits, key=lambda x: x["initial_sum"] * x["rate"] / 100)
profit = max_profit_bank["initial_sum"] * max_profit_bank["rate"] / 100
print(f"  : {max_profit_bank['name']}, : {profit:.2f} .")

# --------------
#  :
#
#   = 3
# _1 50000 5.2
# _2 20000 3.7
# _3 45000 6.8
#   : _2,  : 20000.00 .
#   : _3, : 3060.00 .
```
