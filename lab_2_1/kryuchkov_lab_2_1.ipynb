```python
#  task_02_01_04
# :  . .
# : -231

def is_leap_year(year):
    """,    ."""
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def days_in_month(month, year):
    """    ."""
    days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if month == 2 and is_leap_year(year):
        return 29
    return days[month - 1]

def previous_date(day, month, year):
    """ , ,   ."""
    if day > 1:
        return day - 1, month, year
    else:
        #    
        if month > 1:
            prev_month = month - 1
            prev_day = days_in_month(prev_month, year)
            return prev_day, prev_month, year
        else:
            #    
            return 31, 12, year - 1

def next_date(day, month, year):
    """ , ,   ."""
    if day < days_in_month(month, year):
        return day + 1, month, year
    else:
        #    
        if month < 12:
            return 1, month + 1, year
        else:
            #    
            return 1, 1, year + 1


# 
day, month, year = map(int, input(", ,   : ").split())

# 
prev_d, prev_m, prev_y = previous_date(day, month, year)
next_d, next_m, next_y = next_date(day, month, year)

#    DD/MM/YYYY (  )
print(f" : {prev_d:02d}/{prev_m:02d}/{prev_y}")
print(f" : {next_d:02d}/{next_m:02d}/{next_y}")

# --------------
#  :
#
# , ,   : 1 3 2000
#  : 29/02/2000
#  : 02/03/2000
```

```python
#  task_02_01_28
# :  . .
# : -231

def digital_root_rec(n):
  """    'n' .
    :
        - n (int):   .
    :
        - int:    'n' ( 0  9)."""

  if n < 10:
    return n
  else:
    summa = sum(int(cifra) for cifra in str(n))
    return digital_root_rec(summa)

print(digital_root_rec(65536))
print(digital_root_rec(942))
print(digital_root_rec(8))

# --------------
#  :
#
# 7
# 6
# 8
```

```python
#  task_02_01_29
# :  . .
# : -231

def find_max_rec(numbers):
  """     'numbers' .
    :
        - numbers (list):   .
    :
        - int, float:    ."""

  if len(numbers) == 1:
    return numbers[0]
  else:
    first = numbers[0]
    rest_max = find_max_rec(numbers[1:])
    return first if first > rest_max else rest_max

my_list = [10, 45, 8, 92, 14, 55]
print(find_max_rec(my_list))

# --------------
#  :
#
# 92
```

```python
#  task_02_01_30
# :  . .
# : -231

def calculate_deposit_rec(principal, rate, years):
  """    .
    :
        - principal (float):   .
        - rate (float):    (, 0.05  5%).
        - years (int):  .
    :
        - float:    ."""

  if years == 1:
    return principal * (1 + rate)
  else:
    return principal * ((1 + rate) ** years)

# 1000 ..  8%   5 
final_amount = calculate_deposit_rec(1000, 0.08, 5)
print(f"{final_amount:.2f}")

# --------------
#  :
#
# 1469.33
```
