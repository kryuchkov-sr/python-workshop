```python
# Задание task_02_01_04
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def is_leap_year(year):
    """Проверяет, является ли год високосным."""
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def days_in_month(month, year):
    """Возвращает количество дней в месяце."""
    days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if month == 2 and is_leap_year(year):
        return 29
    return days[month - 1]

def previous_date(day, month, year):
    """Вернуть день, месяц, год предыдущего дня."""
    if day > 1:
        return day - 1, month, year
    else:
        # Переход на предыдущий месяц
        if month > 1:
            prev_month = month - 1
            prev_day = days_in_month(prev_month, year)
            return prev_day, prev_month, year
        else:
            # Переход на предыдущий год
            return 31, 12, year - 1

def next_date(day, month, year):
    """Вернуть день, месяц, год следующего дня."""
    if day < days_in_month(month, year):
        return day + 1, month, year
    else:
        # Переход на следующий месяц
        if month < 12:
            return 1, month + 1, year
        else:
            # Переход на следующий год
            return 1, 1, year + 1


# Ввод
day, month, year = map(int, input("День, месяц, год через пробел: ").split())

# Вычисляем
prev_d, prev_m, prev_y = previous_date(day, month, year)
next_d, next_m, next_y = next_date(day, month, year)

# Вывод в формате DD/MM/YYYY (с ведущими нулями)
print(f"Предыдущий день: {prev_d:02d}/{prev_m:02d}/{prev_y}")
print(f"Следующий день: {next_d:02d}/{next_m:02d}/{next_y}")

# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 3 2000
# Предыдущий день: 29/02/2000
# Следующий день: 02/03/2000
```

```python
# Задание task_02_01_06
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def gcd(a, b):
    while b:
        a, b = b, a % b
    return abs(a)

def lcm(a, b):
    return abs(a * b) // gcd(a, b)

# Ввод
line = input("Введите числа через пробел: ")
nums = list(map(int, line.split()))

# Вычисление
nod = nums[0]
nok = nums[0]

for num in nums[1:]:
    nod = gcd(nod, num)
    nok = lcm(nok, num)

print(f"НОД: {nod}")
print(f"НОК: {nok}")


# Удалите комментарий и допишите код


# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

```python
# Задание task_02_01_16
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def pow1(value, power):
  """Вернуть 'value' в степени 'power'.
    Параметры:
        - !!!.
    Результат:
        !!!."""

  if power == 0:
    return 1
  elif power == 1:
    return value
  else:
    return value ** power

def pow2(value, power):
  """Вернуть 'value' в степени 'power'. Рекурсивный алгоритм.
    Параметры:
        - !!!.
    Результат:
        !!!."""

  if power == 0:
    return 1
  elif power == 1:
    return value
  else:
    return value * pow2(value, power - 1 )

print(pow1(5, 3))
print(pow2(5, 3))


# --------------
# Пример вывода:
#
# 125
# 125
```

```python
# Задание task_02_01_24
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def gcd_rec(a, b):
  """Найти наибольший общий делитель (НОД) чисел 'a' и 'b' рекурсивно.
    Параметры:
        - a (int): первое число.
        - b (int): второе число.
    Результат:
        - int: НОД для a и b."""

  if b == 0:
    return a
  return gcd_rec(b, a % b)

print(gcd_rec(48, 18))
print(gcd_rec(101, 10))

# --------------
# Пример вывода:
#
# 6
# 1
```

```python
# Задание task_02_01_25
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def to_binary_rec(n):
  """Преобразовать десятичное число 'n' в двоичную строку рекурсивно.
    Параметры:
        - n (int): неотрицательное целое число.
    Результат:
        - str: двоичное представление числа 'n'."""

  if n == 0:
    return "0"
  elif n == 1:
    return "1"
  else:
    return to_binary_rec(n // 2) + str(n % 2)

print(to_binary_rec(25))
print(to_binary_rec(7))
print(to_binary_rec(0))

# --------------
# Пример вывода:
#
# 11001
# 111
# 0
```

```python
# Задание task_02_01_28
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def digital_root_rec(n):
  """Найти цифровой корень числа 'n' рекурсивно.
    Параметры:
        - n (int): неотрицательное целое число.
    Результат:
        - int: цифровой корень числа 'n' (от 0 до 9)."""

  if n < 10:
    return n
  else:
    summa = sum(int(cifra) for cifra in str(n))
    return digital_root_rec(summa)

print(digital_root_rec(65536))
print(digital_root_rec(942))
print(digital_root_rec(8))

# --------------
# Пример вывода:
#
# 7
# 6
# 8
```

```python
# Задание task_02_01_29
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def find_max_rec(numbers):
  """Найти максимальный элемент в списке 'numbers' рекурсивно.
    Параметры:
        - numbers (list): непустой список чисел.
    Результат:
        - int, float: максимальное значение в списке."""

  if len(numbers) == 1:
    return numbers[0]
  else:
    first = numbers[0]
    rest_max = find_max_rec(numbers[1:])
    return first if first > rest_max else rest_max

my_list = [10, 45, 8, 92, 14, 55]
print(find_max_rec(my_list))

# --------------
# Пример вывода:
#
# 92
```

```python
# Задание task_02_01_30
# Выполнил: Крючков С. Р.
# Группа: АБП-231

def calculate_deposit_rec(principal, rate, years):
  """Рассчитать итоговую сумму вклада рекурсивно.
    Параметры:
        - principal (float): начальная сумма вклада.
        - rate (float): годовая процентная ставка (например, 0.05 для 5%).
        - years (int): количество лет.
    Результат:
        - float: итоговая сумма на счете."""

  if years == 1:
    return principal * (1 + rate)
  else:
    return principal * ((1 + rate) ** years)

# 1000 у.е. под 8% годовых на 5 лет
final_amount = calculate_deposit_rec(1000, 0.08, 5)
print(f"{final_amount:.2f}")

# --------------
# Пример вывода:
#
# 1469.33
```
